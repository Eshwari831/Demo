```python
# TASK 2: Regression Challenge

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing # search online about this datasets to try new
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
housing=fetch_california_housing()
X= pd.DataFrame(housing.data, columns=housing.feature_names)
y= pd.Series(housing.target, name="Housevalue")

SEED=56
col_to_drop= X.columns[SEED% X.shape[1]]
X=X.drop(col_to_drop,axis=1)
print("dropped column:", col_to_drop)
X_train, X_test, y_train, y_test= train_test_split(
    X, y, test_size=0.3, random_state=SEED)
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
mean=mean_absolute_error(y_test, y_pred)
squ=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(squ)
r2= r2_score(y_test,y_pred)

print("Metrics:\n")
print("MEAN:",round(mean,2)) # central trend of data input
print("SQU:",round(squ,3))
print("RMSE:",round(rmse,1))
print("R2:",round(r2,3))

plt.figure(figsize=(10,6))
plt.scatter(y_test,y_pred,alpha=0.5)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(),y_test.max()],
         'r',linewidth=4)
plt.xlabel("Actual values--")
plt.ylabel("Predicted values--")
plt.title("Actual vs Predicted values\n")
plt.title("signature: Eshwari Sagar Jain\n Roll_no: 10827")
plt.grid(True)
plt.show()
print("Reflection:\n")
print("1] Avgrooms, Avgbedrooms, and population seem to matter most.")
print("2] Model underfits and so much grouping")
print("3] If we drop one column, it didn't break the model as such.")
```
